```python
import csv
df = pd.read_csv('Final_stock_data_V2.csv')
df
```

##### Only adding those that follow the 2/3 criteria into the porfolio models

```python
def check_criteria(each):
    check = []
    if each["capm_status"] == "UnderValued":
        check.append(True)
    if each["Earnings_expectation"] == "Increase":
        check.append(True)
    if each["irv_status"] == "Below Fair Price":
        check.append(True)
    return check
```

```python
def check_number(criteria):
    if len(criteria) >= 2:
        return True
    else:
        return False
```

```python
rows = []
for i in range(938):
    each = df.iloc[i]
    criteria = check_criteria(each)
    result = check_number(criteria)
    if result:
        rows.append(i)
```

```python
len(rows)
```

```python
df2 = df.iloc[rows]
df2
```

```python
#total number of stocks for each sector 

df2["sector"].value_counts()
```

```python
#df for each sector

df_basic_materials = df2.loc[df['sector'] == 'Basic Materials']
df_communication = df2.loc[df['sector'] == 'Communication Services']
df_consumer_c = df2.loc[df['sector'] == 'Consumer Cyclical']
df_consumer_d = df2.loc[df['sector'] == 'Consumer Defensive']
df_energy = df2.loc[df['sector'] == 'Energy']
df_financial = df2.loc[df['sector'] == 'Financial Services']
df_healthcare = df2.loc[df['sector'] == 'Healthcare']
df_industrials = df2.loc[df['sector'] == 'Industrials']
df_real_estate = df2.loc[df['sector'] == 'Real Estate']
df_tech = df2.loc[df['sector'] == 'Technology']
df_utilites = df2.loc[df['sector'] == 'Utilities']
```

##### Filter sectors based on highest returns

```python
#getting each sector's average returns

basic_returns = df_basic_materials['annu_return_5y'].mean()
comm_returns = df_communication['annu_return_5y'].mean()
consumer_c_returns = df_consumer_c['annu_return_5y'].mean()
consumer_d_returns = df_consumer_d['annu_return_5y'].mean()
energy_returns = df_energy['annu_return_5y'].mean()
financial_returns = df_financial['annu_return_5y'].mean()
healthcare_returns = df_healthcare['annu_return_5y'].mean()
industrials_returns = df_industrials['annu_return_5y'].mean()
real_estate_returns = df_real_estate['annu_return_5y'].mean()
tech_returns = df_tech['AHV_5y'].mean()
utilities_returns = df_utilites['AHV_5y'].mean()

print(tabulate([['Basic Materials', basic_returns],['Communication Services', comm_returns],['Consumer Cyclical', consumer_c_returns],['Consumer Defensive', consumer_d_returns], ['Energy', energy_returns], ['Financial Services', financial_returns], ['Healthcare', healthcare_returns], ['Industrials', industrials_returns], ['Real Estate', real_estate_returns], ['Technology', tech_returns], ['Utilities', utilities_returns]],
               headers=['Sector','5 year Annual Returns']))
```

From above, the top 3 sectors that give the highest returns is Technology (0.476139), Utilities (0.260092), Real Estate (0.106064)
